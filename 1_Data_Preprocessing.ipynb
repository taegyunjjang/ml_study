```python
# pandas 라이브러리를 활용한 데이터 전처리 2

abalone_src = base_src+'/abalone.data'
abalone_df = pd.read_csv(base_src+'/abalone.data',
             header=None,sep=',',
             names=['sex','length','diameter','height',
                  'whole_weight','shucked_weight','viscera_weight',
                  'shell_weight','rings'])
```

```python
abalone_df.isnull()
```

```python
abalone_df.isnull().sum()
```

```python
abalone_df.isnull().sum().sum()
```

```python
# 기술통계 확인 => 연속형 변수만 확인 => numerical 변수만 확인 가능
abalone_df.describe()
```

```python
# abalone 성별에 따라 groupby함수를 통해 집계
# ex) DataFrame[집계 변수].groupby(DataFrame[집계 대상])
grouped = abalone_df['whole_weight'].groupby(abalone_df['sex'])
```

```python
grouped
```

```python
grouped.sum()
```

```python
grouped.mean()
```

```python
 # 그룹 변수가 하나가 아닌, 전체 연속형 변수에 대한 집계
 abalone_df.groupby(abalone_df['sex']).mean()
```

```python
# 다음과 같이 간단하게 표현 가능
abalone_df.groupby('sex').mean()
```

```python
# 새로운 조건에 맞는 변수 추가
abalone_df['length_bool'] = np.where(abalone_df['length'] > abalone_df['length'].median(),
                                     'length_long', # True인 경우
                                     'length_short') # False인 경우
```

```python
abalone_df
```

```python
# 그룹 변수를 2개 이상 선택해서 집계
abalone_df.groupby(['sex', 'length_bool']).mean()
```

```python
# 간결한 표현도 가능
abalone_df.groupby(['sex', 'length_bool'])['whole_weight'].mean()
```

```python
# 중복 데이터 삭제(결측치 확인 후 중복된 데이터 확인)
# 중복된 row를 확인하는 법
abalone_df.duplicated().sum()
```

```python
# 중복 예제 데이터 생성
new_abalone = abalone_df.iloc[[0]]
new_abalone_df = pd.concat([abalone_df, new_abalone], axis=0)
new_abalone_df
```

```python
new_abalone_df.duplicated()
```

```python
new_abalone_df.duplicated(keep='last')
```

```python
# 중복 데이터(row 삭제)
new_abalone_df = new_abalone_df.drop_duplicates(keep='last')
```

```python
# NaN(결측치)를 찾아서 다른 값으로 변경
# 기존 데이터에는 결측치가 존재x
abalone_df.isnull().sum().sum()
```

```python
nan_abalone_df = abalone_df.copy()
```

```python
nan_abalone_df.loc[2, 'length'] = np.nan
```

```python
nan_abalone_df
```

```python
# 결측치를 특정 값으로 채우기
zero_abalone_df = nan_abalone_df.fillna(0)
zero_abalone_df
```

```python
# 결측치를 결측치가 속한 컬럼의 평균값으로 대체하기
nan_abalone_df.mean()
```

```python
nan_abalone_df.fillna(nan_abalone_df.mean())
```

```python
# 범주형 변수의 결측치 채우기
nan_abalone_df.loc[2, 'sex'] = np.nan
```

```python
sex_mode = nan_abalone_df['sex'].value_counts()
sex_mode
```

```python
nan_abalone_df['sex'].value_counts().sort_values(ascending=False).index[0]
```

```python
nan_abalone_df['sex'].fillna(sex_mode)
```

```python
# apply()
# 열 기준 집계
abalone_df[['diameter']].apply(np.average, axis=0)
```

```python
# 행 기준 집계
abalone_df[['diameter']].apply(np.average, axis=1)
```

```python
abalone_df[['diameter']].head()
```

```python
# 사용자 함수를 통한 집계
import math
def avg_ceil(x, y, z):
  return math.ceil(x + y + z)

abalone_df[['diameter', 'height', 'whole_weight']].apply(lambda x: avg_ceil(x[0], x[1], x[2]), axis=1)
```

```python
# 문제
answer = abalone_df[['diameter', 'height', 'whole_weight']].apply(lambda x: x[0] + x[1] + x[2] > 1,
                                                         True,
                                                         False)
abalone_df['answer'] = answer
abalone_df
```

```python
# 컬럼 내 unique한 값을 뽑아서 개수 확인 (카테고리 변수)
abalone_df['sex'].value_counts(dropna=True)
```

```python
# 두 개의 DataFrame 합치기 (row)
one_abalone_df = abalone_df.iloc[[0]]
pd.concat([abalone_df, one_abalone_df], axis=0)
```

```python
# 두 개의 DataFrame 합치기 (column)
one_abalone_df = abalone_df.iloc[:, [0]]
pd.concat([abalone_df, one_abalone_df], axis=1)
```
